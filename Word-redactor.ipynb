```python
# pip install pandas
# pip install openpyxl
# pip install -U python-docx
# pip install docxtpl

import pandas as pd
import numpy as np
import docx
from docx import Document
from docxtpl import DocxTemplate
from docx.enum.text import WD_ALIGN_PARAGRAPH
```

## Обозначим файл

```python
doc = Document('ТЗ.docx')
```

```python
# создаем словарь со значениями и строку, которую будет изменять
replace_value = {"обязан": "обязуется",  "и не влиять": "и не влияет", "приложена": "прилагается", "а также достигать": "а также достигает", 
                 " поставлен Заказчику": "поставляется Заказчику", "объекта(- ов)": "объекта (-ов)", "  ": " ",
                 
                "настоящего ": "", "настоящему ": "",

                "должен быть ": "", "должен обеспечивать": "обеспечивает", "должен предоставить": "предоставляет", "должен обеспечить": "обеспечивает",
                "должен включать": "включает", "должен соответствовать": "соответствует", "не должен иметь": "не имеет", 

                "должна сочетаться": "сочетается", "должна обеспечиваться": "обеспечивается", "должна быть ": "", "должна соответствовать": "соответствует",
                "должна соответствовать": "соответствует", "должна осуществляться": "осуществляется", 

                "должно соответствовать": "соответствует", "должно быть подтверждено": "подтверждается", "должно осуществляться": "осуществляется", 
                "не должно создавать": "не создает", 

                "должны проводиться": "проводятся", "должны производиться": "производятся", "должны быть восстановлены": "восстанавливаются", 
                "должны быть оказаны": "оказываются", "должны соответствовать": "соответствуют", "должны сопровождаться": "сопровождаются",
                "должны быть разрешены": "разрешены",

                "не менее чем за 3 (три)": "за 3 (три)", "не позднее чем за 3 (три)": "за 3 (три)", "не менее срока": "равный сроку", 
                "не менее 1 (одного) года": "1 (один) год"
                }
```

```python
# Функция для замены нескольких значений

for i in replace_value:
    for p in doc.paragraphs:
        if p.text.find(i)>=0:
            p.text=p.text.replace(i,replace_value[i])
#save changed document
doc.save('ТЗ_1.docx')
```
